## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x * y * z

<img src="https://raw.githubusercontent.com/gjhernandezp/NM/main/SymPyLab/sympylabwolfram7.jpg" />

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)
x = x1 + I*x2
y = y1 + I*y2
z = z1 + I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```

## What is the average temperature in Arauca in October? make a graph and data

chat.PNG

```
import matplotlib.pyplot as plt

# Data for Arauca's October weather
days = list(range(1, 13))  # Days of rain
temps = [27, 32, 22]  # Average, high, and low temperatures
rainfall = 161  # mm

# Prepare the chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot temperatures
ax1.bar(days, [rainfall / len(days)] * len(days), color='blue', alpha=0.5, label='Rainfall per rainy day (mm)')
ax1.set_xlabel('Rainy Days in October')
ax1.set_ylabel('Rainfall (mm)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Secondary y-axis for temperature data
ax2 = ax1.twinx()
ax2.plot(["Average", "High", "Low"], temps, marker='o', color='red', label='Temperatures (°C)')
ax2.set_ylabel('Temperature (°C)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title and legend
plt.title('Arauca Weather in October')
fig.legend(loc='upper right', bbox_to_anchor=(0.85, 0.85))

plt.tight_layout()
plt.show()
```

```
# prompt: escribir un codigo que me solucione la integral (x^2+5x+2) cot (5x)

import sympy as sp

# Definimos las variables
x = sp.Symbol('x')

# Definimos la función a integrar
func = sp.cot(5*x) * (x**2 + 5*x + 2)

# Intentamos calcular la integral indefinida.  Es probable que no tenga una solución cerrada.
integral = sp.integrate(func, x)

# Mostramos el resultado (que podría ser la integral indefinida o una expresión que indique que no se pudo resolver).
print(integral)

# Si la integral indefinida no se puede calcular simbólicamente, se puede intentar una aproximación numérica
# para un intervalo específico.  Por ejemplo:

# Definimos los límites de integración
a = 0  # Límite inferior
b = 1  # Límite superior

# Calculamos la integral definida numéricamente.
# La función 'sp.integrate' con límites numéricos realiza una integración numérica.
integral_definida = sp.integrate(func, (x, a, b))
print(f"La integral definida de {a} a {b} es aproximadamente: {integral_definida}")

```
